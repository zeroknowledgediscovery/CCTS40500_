```python
import pandas as pd
import glob
import numpy as np
import pylab as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
FNAMES=glob.glob('../top10/*/*7.rnnres')
FLEXROC='~/ZED/Research/cynet_/bin/flexroc '
CF=pd.read_csv('../../cynet_performance/aucvartgt.csv',sep=' ')
CF=CF.round(4)
FNAMES
```

```python
RES=[]
for fname in FNAMES:
    df=pd.read_csv(fname)
    h=np.array(fname.split('/')[2].split('#')).astype(float)
    df[['AR_grt','AR_prd']].to_csv('tmp.csv',header=None,index=None,sep=' ')
    #!cat tmp.csv
    R=! $FLEXROC -i tmp.csv -w 1 -x 0 -t 0.8 -f 0.2 -E 0 -C 1 -L 1
    a=R[0].split()[1]
        
    df[['AS_grt','AS_prd']].to_csv('tmp.csv',header=None,index=None,sep=' ')
    #!cat tmp.csv
    R=! $FLEXROC -i tmp.csv -w 1 -x 0 -t 0.8 -f 0.2 -E 0 -C 1 -L 1
    b=R[0].split()[1]
    
    df[['PR_grt','PR_prd']].to_csv('tmp.csv',header=None,index=None,sep=' ')
    #!cat tmp.csv
    R=! $FLEXROC -i tmp.csv -w 1 -x 0 -t 0.8 -f 0.2 -E 0 -C 1 -L 1
    c=R[0].split()[1]

    res=np.append(h,[a,b,c])
    RES=np.append(RES,res)
    
RES=np.array(RES).astype(float).reshape(int(len(RES)/7),7)
```

```python
rf=pd.DataFrame(RES,columns=['lattgt1','lattgt2','lontgt1','lontgt2',
                             'BURGLARY-THEFT-MOTOR_VEHICLE_THEFT',
                             'HOMICIDE-ASSAULT-BATTERY','VAR'])#.plot()
rf=rf.set_index(['lattgt1','lattgt2','lontgt1','lontgt2']).stack().reset_index().rename(columns={'level_4':'vartgt',0:'auc_NN'})
rf=rf.round(4)
```

```python
mf=CF.set_index(['lattgt1','lattgt2','lontgt1','lontgt2','vartgt'])\
            .join(rf.set_index(['lattgt1','lattgt2','lontgt1','lontgt2','vartgt'])).reset_index().dropna(axis=0)
```

```python
ax=mf.reset_index(drop=True).auc.plot(style='o',label='Cynet')
mf.reset_index(drop=True).auc_NN.plot(ax=ax,style='o',label='NN')
plt.gca().set_ylabel('auc')
plt.gca().legend()
plt.gca().set_xlim(-1,34)
plt.savefig('aucrnn.png',dpi=200,bbox_inches='tight')
```
